```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "resources/schools_complete.csv"
student_data_to_load = "resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
sc = pd.read_csv(school_data_to_load)
st = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(st, sc, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
sc_t = sc["School ID"].count()
st_t = st["Student ID"].count()
budget_t = sc["budget"].sum()
math_avg = st["math_score"].mean()
read_avg = st["reading_score"].mean()

math_pass = df.loc[st["math_score"] >= 70]
math_pct = (math_pass["Student ID"].count() / st_t) * 100

read_pass = df.loc[st["reading_score"] >= 70]
read_pct = (read_pass["Student ID"].count() / st_t) * 100

both_pass = df.loc[(st["math_score"] >= 70) & (df["reading_score"] >= 70)]
both_pct = (both_pass["Student ID"].count() / st_t) * 100

district = {
    "Total Schools": [sc_t],
    "Total Students": [st_t],
    "Total Budget": [budget_t],
    "Average Math Score": [math_avg],
    "Average Reading Score": [read_avg],
    "% Passing Math": [math_pct],
    "% Passing Reading": [read_pct],
    "% Overall Passing": [both_pct]
}

df_dist = pd.DataFrame(district)

df_dist["Total Students"] = df_dist["Total Students"].map("{:,}".format)
df_dist["Total Budget"] = df_dist["Total Budget"].map("${:,.2f}".format)
df_dist["Average Math Score"] = df_dist["Average Math Score"].map("{:.2f}".format)
df_dist["Average Reading Score"] = df_dist["Average Reading Score"].map("{:.2f}".format)
df_dist["% Passing Math"] = df_dist["% Passing Math"].map("{:.2f}".format)
df_dist["% Passing Reading"] = df_dist["% Passing Reading"].map("{:.2f}".format)
df_dist["% Overall Passing"] = df_dist["% Overall Passing"].map("{:.2f}".format)

df_dist.index.name = None

df_dist.style.set_table_styles(
    [{'selector': 'th',
      'props': [('max-width', '70px')]}]
)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
df_group = df.groupby(["school_name"])

sc_type = df_group["type"].first()
st_t = df_group["Student ID"].count()
budget_t = df_group["budget"].first()
budget_avg = budget_t / st_t
math_avg = df_group["math_score"].mean().astype(float)
read_avg = df_group["reading_score"].mean().astype(float)

math = df.loc[df["math_score"] >= 70]
df_math = math.groupby(["school_name"])
math_pct = (df_math["Student ID"].count() / st_t) * 100

read = df.loc[df["reading_score"] >= 70]
df_read = read.groupby(["school_name"])
read_pct = (df_read["Student ID"].count() / st_t) * 100

both = df.loc[(df["math_score"] >= 70) & (df["reading_score"] >= 70)]
df_both = both.groupby(["school_name"])
both_pct = (df_both["Student ID"].count() / st_t) * 100

school = {
    "School Type": sc_type,
    "Total Students": st_t,
    "Total School Budget": budget_t,
    "Per Student Budget":budget_avg,
    "Average Math Score": math_avg,
    "Average Reading Score": read_avg,
    "% Passing Math": math_pct,
    "% Passing Reading": read_pct,
    "% Overall Passing": both_pct
}

df_sch = pd.DataFrame(school)

df_sch["Total Students"] = df_sch["Total Students"].map("{:,}".format)
df_sch["Total School Budget"] = df_sch["Total School Budget"].map("${:,.2f}".format)
df_sch["Per Student Budget"] = df_sch["Per Student Budget"].map("${:,.2f}".format)
df_sch["Average Math Score"] = df_sch["Average Math Score"].map("{:.2f}".format)
df_sch["Average Reading Score"] = df_sch["Average Reading Score"].map("{:.2f}".format)
df_sch["% Passing Math"] = df_sch["% Passing Math"].map("{:.2f}".format)
df_sch["% Passing Reading"] = df_sch["% Passing Reading"].map("{:.2f}".format)
df_sch["% Overall Passing"] = df_sch["% Overall Passing"].map("{:.2f}".format)

df_sch.index.name = None

df_sch.style.set_table_styles(
    [{'selector': 'th',
      'props': [('max-width', '70px')]}]
)
```

## Math Scores by Grade

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
df_spend = pd.DataFrame(df_sch)

df_spend["Per Student Budget"] = df_spend["Per Student Budget"].str.replace('$', '')

min_b = float(df_spend["Per Student Budget"].min())
max_b = float(df_spend["Per Student Budget"].max())

delta = int((max_b - min_b) / 3)
range1 = int(min_b) + delta
range2 = int(range1) + delta

min_g = "<$" + str(int(min_b))
range1_g = "$" + str(int(min_b)+1) + "-" + str(range1)
range2_g = "$" + str(int(range1)+1) + "-" + str(range2)
max_g = "$" + str(int(range2)+1) + "-" + str(int(max_b))

# Bins defined by calculated logic above:
bins = [0, int(min_b), range1, range2, int(max_b)]
spend_group = [min_g,range1_g,range2_g,max_g]

# Bins defined by instructions received:
#bins = [0, 584, 629, 644, 675]
#spend_group = ["<584", "585-629", "630-644", "645-675"]

df_spend_bin = pd.DataFrame(df_spend)
df_spend_bin["Per Student Budget"] = df_spend_bin["Per Student Budget"].astype(float)
df_spend_bin["Spending Group"] = pd.cut(df_spend_bin["Per Student Budget"], bins, labels=spend_group)

df_spend["Per Student Budget"] = "$" + df_spend["Per Student Budget"]

df_spend_group = df_spend_bin.groupby("Spending Group")

def getavg(lst):
    lst_avg = []
    for i in lst:
        if len(i[1]) > 0:
            sum1=0
            average=0 
            for j in i[1]:
                sum1 += float(j)                 
                average = round(sum1/len(i[1]),2)
            lst_avg.append(average)
    return lst_avg

score = {
    "Average Math Score": getavg(df_spend_group["Average Math Score"]),
    "Average Reading Score": getavg(df_spend_group["Average Reading Score"]),
    "% Passing Math": getavg(df_spend_group["% Passing Math"]),
    "% Passing Reading": getavg(df_spend_group["% Passing Reading"]),
    "% Overall Passing": getavg(df_spend_group["% Overall Passing"])
}

df_spend_score = pd.DataFrame(score)
df_spend_score.index = [str(spend_group[i]) for i in range(0, len(spend_group))]
df_spend_score.index.name = "Spending Ranges (Per Student)"
df_spend_score.head()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Bins defined by instructions received:
bins = [0, 999, 1999, 4999]
size_group = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

df_size_bin = pd.DataFrame(df_sch)
df_size_bin["Total Students"] = df_size_bin["Total Students"].str.replace(',', '')
df_size_bin["School Size"] = pd.cut(df_size_bin["Total Students"].astype(int), bins, labels=size_group)

df_size_group = df_size_bin.groupby("School Size")

sch_size = {
    "Average Math Score": getavg(df_size_group["Average Math Score"]),
    "Average Reading Score": getavg(df_size_group["Average Reading Score"]),
    "% Passing Math": getavg(df_size_group["% Passing Math"]),
    "% Passing Reading": getavg(df_size_group["% Passing Reading"]),
    "% Overall Passing": getavg(df_size_group["% Overall Passing"])
}

df_size_score = pd.DataFrame(sch_size)
df_size_score.index = [str(size_group[i]) for i in range(0, len(size_group))]
df_size_score.index.name = "School Size"
df_size_score.head()
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
# Bins defined by instructions received:
bins = [-1,0,1]
type_group = ["Charter", "District"]

df_type_bin = pd.DataFrame(df_sch)
lst_int = []

for x in df_type_bin["School Type"]:
    if x == "Charter":
        lst_int.append(0)
    elif x == "District":
        lst_int.append(1)

df_type_bin["School Type"] = pd.cut(lst_int, bins, labels=type_group)

df_type_group = df_type_bin.groupby("School Type")

sch_type = {
    "Average Math Score": getavg(df_type_group["Average Math Score"]),
    "Average Reading Score": getavg(df_type_group["Average Reading Score"]),
    "% Passing Math": getavg(df_type_group["% Passing Math"]),
    "% Passing Reading": getavg(df_type_group["% Passing Reading"]),
    "% Overall Passing": getavg(df_type_group["% Overall Passing"])
}

df_type_score = pd.DataFrame(sch_type)
df_type_score.index = [str(type_group[i]) for i in range(0, len(type_group))]
df_type_score.index.name = "School Type"
df_type_score.head()
```
